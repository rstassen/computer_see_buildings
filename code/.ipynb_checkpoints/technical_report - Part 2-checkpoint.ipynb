```python
# importing libraries

import numpy as np, pandas as pd
import segmentation_models as sm
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import glob
import os

from sklearn.model_selection import train_test_split
from keras import optimizers
from PIL import Image
```

```python
# # this is needed in some versions of tensorflow/keras
# sm.set_framework('tf.keras')
# sm.framework()
```

```python
# setting preprocessor
BACKBONE = 'resnet34'
preprocessor = sm.get_preprocessing(BACKBONE)
```

```python
# importing data
from google.colab import drive

# connecting to personal drive; will trigger authorization request
drive.mount('/content/drive/')
```

```python
# appending training pngs to empty list
train_images = []

for i in range(3851):
  img = cv2.imread(f'/content/drive/MyDrive/GA_capstone/images_masks/images/image_{i}.png')
  img = cv2.resize(img, (256,256))
  train_images.append(img)

train_images = np.array(train_images)
```

```python
# appending training masks to empty list
train_masks = []

for i in range(3851):
  img = cv2.imread(f'/content/drive/MyDrive/GA_capstone/images_masks/masks/mask_{i}.png', 0)
  img = cv2.resize(img, (256,256))
  train_masks.append(img)

train_masks = np.array(train_masks)
```

```python
# confirming proper imports

fig = plt.figure(figsize=(15,15))

fig.add_subplot(1,2,1)
plt.imshow(train_images[96])
plt.title("Sample Training Image")

fig.add_subplot(1,2,2)
plt.imshow(train_masks[96])
plt.title("Sample Mask")

plt.show();
```

```python
# checking data shape

X = train_images
y = train_masks
print(X.shape)
print(y.shape)

print()

# expanding dimensions for CNN input
y = np.expand_dims(y, axis=3)

print(X.shape)
print(y.shape)
```

```python
# train-test splits
X_train, X_val, y_train, y_val = train_test_split(X,y, train_size=0.75, random_state=42)

print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
```

```python
# preprocess inputs
X_train = preprocessor(X_train)
X_val = preprocessor(X_val)
```

```python
# confirming GPU in use
tf.test.gpu_device_name()
```

```python
# instantiate optimizer
opt_adam = optimizers.Adam(learning_rate=0.00005)
# opt_sgd = optimizers.SGD(learning_rate=0.0001) # adam performed better than SGD in my tests
```

```python
# defining model
model = sm.Unet(BACKBONE, encoder_weights='imagenet')
model.compile(optimizer=opt_adam, loss=sm.losses.bce_jaccard_loss, metrics=[sm.metrics.iou_score])

print(model.summary())
```

```python
# training model
history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    batch_size=32,
                    epochs=20,
                    verbose=1)
```

```python
# plotting loss trend, consistently downward

loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss)+1)

plt.figure(figsize=(8,6))
plt.plot(epochs, loss, 'y', label = "Training Loss")
plt.plot(epochs, val_loss, 'r', label = "Validation Loss")
plt.title("Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# model overfit is evident as testing_iou score hits a wall
score = history.history['iou_score']
val_score = history.history['val_iou_score']
epochs = range(1, len(loss)+1)

plt.figure(figsize=(8,6))
plt.plot(epochs, score, 'y', label = "Training IoU Score")
plt.plot(epochs, val_score, 'r', label = "Validation IoU Score")
plt.title("Training vs Validation IoU Scores")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

## Model Testing

```python
# function to format testing images

def image_transform(img_path):

    # importing image as grayscale
    img = cv2.imread(img_path, 0)

    # crop image
    x = img.shape[0]
    y = img.shape[1]
    cropped_img = img[0:min(x,y), 0:min(x,y)]

    # resize image
    resize_img = cv2.resize(cropped_img, (256,256))

    # transform image to 3-channel grayscale, to satisfy model's dimension requirements
    # this code was modified from Mona Jalal
    # https://askubuntu.com/questions/1091493/convert-a-1-channel-image-to-a-3-channel-image
    img_out = np.zeros((256,256,3), dtype='int')
    img_out[:,:,0] = resize_img
    img_out[:,:,1] = resize_img
    img_out[:,:,2] = resize_img
    
    return img_out
```
